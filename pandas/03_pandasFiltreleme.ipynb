### **Pandas Filtreleme**

```python
# Pandas kütüphanesini içe aktralım.
import pandas as pd
```

```python
# ad,soyad,meslek ve maasla ilgili listeler oluşturalım ve bunları sözlüğe aktaralım.
ad = ["Ali","Halit","İbrahim","Selin","Selim","Rüya","Bervin","Merve"]
soyad = ["Kelepir","Ozan","Rona","Çelik","Büyüm","Ayhan","Çiçek","Nur"]
meslek = ["Öğretmen","Bilgisayar Mühendisi","Öğretmen","Hemşire","Memur","Doktor","Siber Güvenlik Uzmanı","Hemşire"]
maas = [5000,7000,5000,5500,4000,10000,7000,6000]
dictionary = {"ad" : ad,
"soyad" : soyad,
"meslek" : meslek,
"maas": maas
}
data = pd.DataFrame(dictionary)
data
```

```python
# ad, soyad sütunlarını birleştirelim ve birleştirdiğimiz bu sütunlar ile yeni bir sütun oluşturup yeniden adlandıralım.
data["ad ve soyad"] = data.ad +" "+ data.soyad
data
```

```python
# Adının ilk harfi "S" olanları bulalım.
filterName = data.ad.str.startswith("S")
data[filterName]
```

```python
# Adının son iki harfi "im" olanları bulalım.
filterName2 = data.ad.str.endswith("im")
data[filterName2]
```

```python
# Adının herhangi bir yerinde "l" harfi olanları bulalım.
filterName3 = data.ad.str.contains("l")
data[filterName3]
```

```python
# Adının herhangi bir yerinde "l" harfi olanları bulalım.
filterName3 = data.ad.str.contains("l")
data[filterName3]
```

```python
# maas 5000 den büyük olanları filtreleyelim.
filtre1 = data.maas > 5000
data[filtre1]
```

```python
# maas'ı 4000 den büyük ve mesleği Hemşire olanları bulalım.
filtre2 = data.meslek == "Hemşire"
data[filtre1 & filtre2]
```

```python
# Öğretmenlerin toplam sayısını bulalım.
filterCount = data.meslek.str.count("Öğretmen")
sum(filterCount)
```

#### **Pandas Liste Kavrama**

```python
# Pandas kütüphanesini içe aktralım.
import pandas as pd
```

```python
dictionary2 = {"ögreciler" : ["Selim","Kerem","Halit","Mert","Ayşe","Hilal"],
"bölüm" : ["YBS","Muhasebe","İşletme","İktisat","Maliye","YBS"],
"vize" : [50,70,80,85,55,42],
"final": [80,20,100,75,66,95]
}
data2 = pd.DataFrame(dictionary2)
data2
```

```python
# Öğrencilerin vize ve final notlarının ortalamasını alalım.
data2["vize_final_ort"] = data2["vize"]*0.4 + data2["final"]*0.6
data2
```

```python
# Sınıf ortalamasını bulalım.
meanS = data2.vize_final_ort.mean()
print(f"Sınıf ortalaması: {meanS}")
```

```python
# Sınıf ortalamasını bulduktan sonra verimizi yüksek ve düşük olmak üzere ikiye ayıralım.
data2["BAŞARI_DURUMU"] = ["düşük" if meanS > i else "yüksek" for i in data2.vize_final_ort]
data2
```

```python
# Başka bir örnek ise verimizdeki sütun isimlerinin hepsini küçük harfe uyarlayabiliriz. BAŞARI_DURUMU küçük harfe uyarladık.
data2.columns = [i.lower() for i in data2.columns]
data2
```

```python
# Liste kavrama yöntemin de Lambda fonksiyonu kullanılabilir.
# assign metodu atama yapar.
data3 = data2.assign(ortalama_5_puan_eklenmis_hali = lambda x: (x["vize_final_ort"] + 5) ) 
data3
```

```python
# apply yani uygula metodu lambda yerine kullanılabilir.
def ort10Puan(puan):
    return puan + 10
data3["ortalama_10_puan_eklenmis_hali"] = data3.vize_final_ort.apply(ort10Puan)
data3
```
